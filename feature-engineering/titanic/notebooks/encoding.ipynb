```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder

df=pd.read_csv("../data/train.csv")
df
```

```python
new_df=df.copy()
new_df.drop("Cabin", axis=1, inplace=True)
new_df
```

```python
new_df.columns
```

```python
new_df["Embarked"].fillna(new_df["Embarked"].mode()[0], inplace=True)
new_df.isnull().sum()
```

```python
#one hot encoding
new_df = pd.get_dummies(new_df, columns=["Embarked"], drop_first=True) # S=0.0 C=1.0 Q=2.0
```

```python
new_df["Age"] = new_df["Age"].fillna(new_df["Age"].median()) 
new_df.isnull().sum()
```

```python
#Label Encoding 
label_encoder=LabelEncoder()
new_df["Sex"]=label_encoder.fit_transform(new_df["Sex"]) # male:1 female:0
new_df
```

```python
new_df["FamilySize"] = new_df["SibSp"] + new_df["Parch"] + 1
new_df
```

```python
new_df.drop('Ticket', axis=1, inplace=True)
new_df
```
